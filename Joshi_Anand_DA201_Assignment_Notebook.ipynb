```python
# My GitHub repository.


# Import necessary pandas
import pandas as pd
import numpy as np

# Read the CSV file into a pandas DataFrame
ad = pd.read_csv('actual_duration.csv')

# Read the CSV file into a pandas DataFrame
ar = pd.read_csv('appointments_regional.csv')

# Read the Excel file into a pandas DataFrame
nc = pd.read_excel('national_categories.xlsx') 

# Display column names, number of rows and columns, data types, and missing values for for ad DataFrame
print("actual_duration.csv DataFrame:")
print(ad.columns)
print(ad.shape)
print(ad.dtypes)
print(ad.isnull().sum())

# Display column names, number of rows and columns, data types, and missing values for ar DataFrame
print("\nappointments_regional.csv DataFrame:")
print(ar.columns)
print(ar.shape)
print(ar.dtypes)
print(ar.isnull().sum())

# Display column names, number of rows and columns, data types, and missing values for nc DataFrame
print("\nnational_categories.xlsx DataFrame:")
print(nc.columns)
print(nc.shape)
print(nc.dtypes)
print(nc.isnull().sum()) 
```

```python
# Read the metadata file
with open('metadata_nhs.txt', 'r') as f:
    metadata = f.read()
    
print(metadata)
```

```python
# Display descriptive statistics and metadata for ad DataFrame
print("actual_duration.csv DataFrame:")
print(ad.describe())
print(ad.info())

# Display descriptive statistics and metadata for ar DataFrame
print("\nappointments_regional.csv DataFrame:")
print(ar.describe())
print(ar.info())

# Display descriptive statistics and metadata for nc DataFrame
print("\nnational_categories.xlsx DataFrame:")
print(nc.describe())
print(nc.info())
```
